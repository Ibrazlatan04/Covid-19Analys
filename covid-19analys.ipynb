# Covid-19 dataanalys

Här har jag gjort en arbetsbok där jag beskivit hur sjukdomspridningen sett ut i Europa och Sverige 2022.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")
df

```

## Filtering av data

```python
df_1 = df[df['countriesAndTerritories'] == 'Sweden'] 
df_1
```

```python
df_2 = df_1 [df_1['year'] == 2022]
df_2
```

```python
df_3 = df_2[df_2['day'] == 17]
df_3
```

```python
df_filtering = df[(df['year'] == 2022) & (df['day'] == 17) & (df['countriesAndTerritories'] == 'Sweden')]
df_filtering
```

## Visualisering av data

```python
plt.plot (df_3["month"], df_3["cases"], 'ro--')
plt.title ('Antal fall av covid den 17 under 2022')
```
